```python
import sqlite3 as sq
import pandas as pd
con = sq.connect("/content/netflix.sqlite")
```

```python
cur = con.cursor()
```

```python
with con:
  cur.execute(""" CREATE TABLE actors (
    id INT NOT NULL PRIMARY KEY,
    actor_name TEXT
  );
   """)
```

```python
cur.execute(""" INSERT INTO actors (id, actor_name)
SELECT show_id, "cast"
FROM netflix_titles;
   """)
```

```python
pd.read_sql('''
SELECT *
FROM actors
''', con)
```

```python
with con:
  cur.execute(""" CREATE TABLE films (
    id INT NOT NULL PRIMARY KEY,
    film_title TEXT
  );
   """)
```

```python
cur.execute(""" INSERT INTO films (id, film_title)
SELECT show_id, title
FROM netflix_titles;
   """)
```

```python
pd.read_sql('''
SELECT *
FROM films
''', con)
```

```python
with con:
  cur.execute(""" CREATE TABLE films_actors (
    id INT NOT NULL PRIMARY KEY,
    actors TEXT,
    film_title TEXT
  );
   """)
```

```python
cur.execute(""" INSERT INTO films_actors (id, actors, film_title)
SELECT films.id, actors.actor_name, films.film_title
FROM films, actors
WHERE(films.id = actors.id)
   ;""")
```

```python
pd.read_sql('''
SELECT *
FROM films_actors
''', con)
```
